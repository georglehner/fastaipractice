# A simple fruit classifier

We trained a model using the fast.ai package with Bing image search data to be able to recognize different sorts of fruits. At the moment it works on apples, bananas, oranges, raspberries, avocados and passion fruits. Requests for more fruit types are always welcome.

Upload a picture of your fruit and it will tell you which fruit it is!

```python
path = Path()
learn_inf = load_learner(path/'export.pkl')
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
def on_click(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
```

```python
btn_upload.observe(on_click, names=['data'])
```

```python
display(VBox([widgets.Label('Select your fruit!'), 
      btn_upload, out_pl, lbl_pred]))
```
